## Contagem de Palavras: *Moby Dick*
###### by Herman Melville

```
import requests
from bs4 import BeautifulSoup
import nltk
```

```
r = requests.get("https://www.gutenberg.org/files/2701/2701-h/2701-h.htm")

r.encoding = 'utf-8'

html = r.text
```

```
soup = BeautifulSoup(html, 'html.parser')

text = soup.get_text()

text = text[32270:-19040]

print(text[:1000])
```

```
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')

tokens = tokenizer.tokenize(text)

print(tokens[0:8])
```

```
words = []

for word in tokens:
  words.append(word.lower())

print(words[0:8])
```

```
from nltk.corpus import stopwords
nltk.download('stopwords')
sw = stopwords.words('english')
```

```
words_ns =  [word for word in words if word not in sw]
```

```
print(words_ns[0:8])
```

```
freqdist = nltk.FreqDist(words_ns)
```

```
freqdist
```

```
%matplotlib inline

freqdist.plot(500)
```

## Contagem de Palavras: *Alice’s Adventures in Wonderland*
###### by Lewis Carroll

```
r = requests.get("https://gutenberg.org/cache/epub/11/pg11-images.html")

r.encoding = 'utf-8'

html = r.text
```

```
soup = BeautifulSoup(html, 'html.parser')

text = soup.get_text()
```

```
text = text[4132:-18362]
print(text[:1000])
```

```
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')

tokens = tokenizer.tokenize(text)

print(tokens[0:8])
```

```
words = [word.lower() for word in tokens]

print(words[0:8])
```

```
words_ns =  [word for word in words if word not in sw]

print(words_ns[0:8])
```

```
freqdist = nltk.FreqDist(words_ns)
```

```
freqdist
```

```
%matplotlib inline

freqdist.plot(25)
```

No gráfico acima podemos ver as 25 palavras mais frequentes no livro, que realmente estão relacionadas com o enredo do livro *Alice's Adventures in Wonderland*, e já é possível verificar que o gráfico satisfaz a lei de Zipf. Na próxima célula, em um plot com mais palavras, conseguimos confirmar a lei de Zipf.

```
freqdist.plot(500)
```

```
# Bigramas

from nltk import bigrams

string_bigrams = list(bigrams(tokens))
print(string_bigrams)
```
